Task 1: Conceptual Understanding (Written Answers)
1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" can be treated as two different words if we don’t convert the text into lowercase. Even though both words mean the same thing for humans, computers see them as separate tokens. This can increase the vocabulary size unnecessarily and may reduce the accuracy of NLP models. That’s why lowercase conversion is usually done during preprocessing.

2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many very common words like "is", "the", "and", "in", which appear frequently but do not add much meaning. These words can create noise, increase processing time, and may reduce the performance of machine learning models because the model may focus more on these common words instead of meaningful keywords.

3. Provide two real-world scenarios where removing stopwords can be harmful

Removing stopwords is not always a good idea because sometimes stopwords carry important meaning.

Scenario 1: Sentiment Analysis
Words like "not", "no", "never" are often considered stopwords, but they completely change the meaning of a sentence.
Example:
"I am not satisfied"
If "not" is removed, the sentence becomes "I am satisfied", which gives the wrong sentiment.

Scenario 2: Search Engines or Question Answering Systems
In systems like Google search or chatbots, words like "to", "from", "where", "how" help understand the user’s query.
Example:
"How to apply for internship?"
If stopwords are removed, the sentence becomes unclear and the system may not understand the intent properly.

4. Explain the difference between stemming and lemmatization

Stemming is a basic technique where we remove the ending of a word to get its root form. It is fast, but the root word may not be a real dictionary word.
Example:
"studying" -> "studi"

Lemmatization is a more advanced method where the word is converted into its proper dictionary form using grammar rules. It gives more meaningful results compared to stemming.
Example:
"studying" -> "study"

So, stemming is faster but less accurate, while lemmatization is more accurate but takes more processing time.

```python
import re

def preprocess_text(text):
    try:
        # 1. Handle None or empty input
        if text is None:
            return ""

        # Convert to string (if not already)
        text = str(text)

        # 2. Convert to lowercase
        text = text.lower()

        # 3. Remove URLs (http, https, www)
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        # 4. Remove email-like patterns
        text = re.sub(r'\S+@\S+', '', text)

        # 5. Remove numbers
        text = re.sub(r'\d+', '', text)

        # 6. Handle repeated characters (soooo -> so)
        # Replace 3+ repeated characters with 1 character
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # 7. Remove punctuation/special symbols (keep only letters and spaces)
        text = re.sub(r'[^a-z\s]', '', text)

        # 8. Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # 9. Remove very short tokens (length <= 2)
        # Exception: keep "no" and "not"
        words = text.split()
        cleaned_words = []

        for word in words:
            if len(word) > 2 or word in ["no", "not"]:
                cleaned_words.append(word)

        # Join words back
        text = " ".join(cleaned_words)

        return text

    except Exception as e:
        return ""
```

```python
import re

def preprocess_text(text):
    try:
        # Handle None
        if text is None:
            return ""

        # Convert to string
        text = str(text).strip()

        # Handle empty string
        if text == "":
            return ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        # Remove emails
        text = re.sub(r'\S+@\S+', '', text)

        # Remove numbers
        text = re.sub(r'\d+', '', text)

        # Handle repeated characters (soooo -> so)
        text = re.sub(r'(.)\1{2,}', r'\1', text)

        # Remove special characters (emojis, punctuation)
        text = re.sub(r'[^a-z\s]', '', text)

        # Remove extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        # If after cleaning nothing remains
        if text == "":
            return ""

        # Remove short tokens (<=2) but keep no and not
        words = text.split()
        cleaned_words = []

        for word in words:
            if len(word) > 2 or word in ["no", "not"]:
                cleaned_words.append(word)

        return " ".join(cleaned_words)

    except Exception:
        return ""


def full_pipeline(text_list):
    try:
        if text_list is None or len(text_list) == 0:
            return {"tokens": [], "clean_sentences": []}

        clean_sentences = []
        all_tokens = []

        for text in text_list:
            cleaned = preprocess_text(text)
            clean_sentences.append(cleaned)

            tokens = cleaned.split()
            all_tokens.extend(tokens)

        return {
            "tokens": all_tokens,
            "clean_sentences": clean_sentences
        }

    except Exception:
        return {"tokens": [], "clean_sentences": []}


# ---------------- TASK 7 TEST CASES ----------------

test_inputs = ["", "😍😍😂😂", "1234567890"]

result = full_pipeline(test_inputs)

print("Original Inputs:", test_inputs)
print("Clean Sentences:", result["clean_sentences"])
print("Tokens:", result["tokens"])
```
